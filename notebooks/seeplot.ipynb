```python
import torch
import numpy as np
import matplotlib.pyplot as plt
```

```python
ckpt_path = "/Users/raunavghosh/Documents/code_projects/ToyGenerativeModels/checkpoint.pt"
```

```python
checkpoint = torch.load(ckpt_path, map_location="cpu")
```

```python
checkpoint.keys()
```

```python
smooth_iter = 100
train_loss = checkpoint['train_loss']
x = len(train_loss)
x = x - (x % smooth_iter)
train_loss = np.reshape(train_loss[:-1], (-1, smooth_iter))
train_loss = np.log10(train_loss)
plt.plot(train_loss.reshape(-1))
train_loss = np.mean(train_loss, axis=1)
plt.plot(np.linspace(0, x, len(train_loss)), train_loss)
plt.legend(['Noisy loss', "Smooth loss"])
plt.ylabel("log scale")
plt.xlabel("Steps")
plt.show()
```
